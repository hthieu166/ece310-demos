# Computing and plotting the DTFT

```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
pi = np.pi
```

## DTFT function

```python
def DTFT(x, n0 = 0, w = np.linspace(-pi,pi,100)):
    '''
    DTFT    Compute the Discrete-Time Fourier Transform of a signal
        Input:
          x  : discrete-time signal in an array
          n0 : index of the signal corresponding to the sample x[0]
          w  : array of frequencies that we want to compute DTFT
               for example: w = np.linspace(-pi,pi,100)
        Output:
          X  : DTFT coefficients at the specified frequencies w
    '''
    X = np.zeros_like(w,dtype=complex) # data types: important in Python!!
    # For each frequency in w
    for k in range(len(w)):
        # Compute the DTFT sum over signal samples
        X[k] = 0
        for n in range(len(x)):
            X[k] = X[k] + x[n] * np.exp(-1j * w[k] * (n - n0))

    # Plot output
    plt.subplot(1,2,1)    
    plt.plot(w/pi, np.absolute(X))
    plt.xlabel('$\omega/\pi$')
    plt.title('$|X_d(\omega)|$'); # Plot out the DTFT in magnitude
    
    plt.subplot(1,2,2)    
    plt.plot(w/pi, np.angle(X))
    plt.xlabel('$\omega/\pi$')
    plt.title('angle $X_d(\omega)$'); # Plot out the DTFT in angle (phase)
    
    return X
```

```python
# Frequencies to evaluate the DTFT
# 100 linearly equally spaced points between -pi and pi
w = np.linspace(-pi, pi, 100)
```

## First example signal

```python
x = np.array([1, 1])
X = DTFT(x, 0, w) # Compute the DTFT of x
```

## Second example signal

```python
x = np.array([1, -1])
X = DTFT(x, 0, w) # Compute the DTFT of x
```
