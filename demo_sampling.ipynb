# Sampling

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Continuous-domain sinusoidal signal
f = 30  # frequency in cycles per second (or Hz)
W = 2*np.pi*f # equivalent frequency in radiants per second
S = 0.1 

t = np.linspace(0, S, 1000) # continuous time t in [0, S]
xc = np.cos(W*t)

plt.plot(t, xc)
plt.title('$x_c(t)$')
plt.xlabel('$t$')
```

```python
fs = 100 # sampling frequency
T = 1/fs # equivalent sampling interval

# Discrete-domain sinusoidal signal
n = np.arange(S/T)
x = np.cos(W*n*T)

plt.stem(x)
plt.title('$x[n]$')
plt.xlabel('$n$')
```

```python
# Plot both continuous-domain and discrete-domain signals on the same plot
plt.plot(t, xc)
plt.stem(n*T, x, 'g')  # index n is multiplied by T

plt.title('$x_c(t), x[n]$')
plt.xlabel('$t$')
```

```python
# Alias frequency
fa = fs - f
Wa = 2*np.pi*fa
xa = np.cos(Wa*t)

plt.plot(t, xc)
plt.plot(t, xa)
plt.stem(n*T, x, 'g')  # index n is multiplied by T

plt.title('$x_c(t), x_{a}(t), x[n]$')
plt.xlabel('$t$')
```

```python
# Next alias frequency
fa = fs + f
Wa = 2*np.pi*fa
xa = np.cos(Wa*t)

plt.plot(t, xc)
plt.plot(t, xa)
plt.stem(n*T, x, 'g')  # index n is multiplied by T

plt.title('$x_c(t), x_{a}(t), x[n]$')
plt.xlabel('$t$')
```

```python
# Next alias frequency
fa = 2*fs - f
Wa = 2*np.pi*fa
xa = np.cos(Wa*t)

plt.plot(t, xc)
plt.plot(t, xa)
plt.stem(n*T, x, 'g')  # index n is multiplied by T

plt.title('$x_c(t), x_{a}(t), x[n]$')
plt.xlabel('$t$')
```
