# Convolution

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
from scipy.signal import convolve
```

## Convolution with simple signals
### Impulse
$ y[n] = \delta[n] * h[n] = h[n]$

```python
from scipy.signal import unit_impulse
```

```python
N = 8 # length of x
M = 8 # length of h
L = N + M - 1 # length of y
n = np.arange(N)
m = np.arange(M)
l = np.arange(L)

x = unit_impulse(N,0)
h = n # h is the ramp function
y = convolve(x,h)

plt.figure(figsize=(10,15))
plt.subplot(311)
plt.stem(n,x)
plt.title('x')
plt.subplot(312)
plt.stem(m,h)
plt.title('h')
plt.subplot(313)
plt.stem(l,y)
plt.title('y = x * h')
plt.show()
```

## Shifted impulse
$y[n] = \delta[n - n_0] * h[n] = h[n - n_0]$

```python
N = 8 # length of x
M = 8 # length of h
L = N + M - 1 # length of y
n = np.arange(N)
m = np.arange(M)
l = np.arange(L)

x = unit_impulse(N,5) # change this number to change the offset, n_0
h = n # x is the ramp function
y = convolve(x,h)

plt.figure(figsize=(10,15))
plt.subplot(311)
plt.stem(n,x)
plt.title('x')
plt.subplot(312)
plt.stem(m,h)
plt.title('h')
plt.subplot(313)
plt.stem(l,y)
plt.title('y = x * h')
plt.show()
```

## Convolution of two signals
$ x[n] = \begin{cases} 1 & 0 \leq n < 2  \\ 0 & else \end{cases} $

$ h[n] = \begin{cases} n & 0 \leq n < 4  \\ 0 & else \end{cases} $

$ y[n] = x[n] * h[n] $

```python
N = 8 # length of x
M = 8 # length of h
L = N + M - 1 # length of y
n = np.arange(N)
m = np.arange(M)
l = np.arange(L)

x = np.zeros(M)
x[0:2] = np.array([1, 1])
h = np.zeros(N)
h[0:4] = np.array([0, 1, 2, 3])

y = convolve(x, h)
plt.figure(figsize=(10,15))
plt.subplot(311)
plt.stem(n, x)
plt.title('x')
plt.subplot(312)
plt.stem(m, h)
plt.title('h')
plt.subplot(313)
plt.stem(l, y)
plt.title('y = x * h')
plt.show()
```

## Demonstrating linear and time-invariant system is convolution

Why we can use the **impulse response** to characterize a system

```python
delta0 = unit_impulse(N,0)
delta1 = unit_impulse(N,1)

y0 = convolve(delta0, h)
y1 = convolve(delta1, h)

y = x[0]*y0 + x[1]*y1

plt.figure(figsize=(10,15))
plt.subplot(311)
plt.stem(l,y0)
plt.title('$\delta[n] * h[n] = h[n]$')
plt.subplot(312)
plt.stem(l,y1)
plt.title('$\delta[n-1] * h[n] = h[n-1]$')
plt.subplot(313)
plt.stem(l,y)
plt.title('Linear combination of all impulse convolutions')
plt.show()
```
